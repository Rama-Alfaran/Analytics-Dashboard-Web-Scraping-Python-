```python
%%writefile /content/drive/MyDrive/vendor-analysis-project/database/models.py
"""
Database Models
Define database schema using SQLAlchemy ORM
"""
from sqlalchemy import create_engine, Column, Integer, String, Float, Date, Boolean, DateTime, Text, ForeignKey
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship
from datetime import datetime

Base = declarative_base()

class Vendor(Base):
    """Vendor Master Table"""
    __tablename__ = 'vendors'

    vendor_id = Column(Integer, primary_key=True, autoincrement=True)
    vendor_code = Column(String(50), unique=True, nullable=False)
    vendor_name = Column(String(255), nullable=False)
    email = Column(String(255))
    phone = Column(String(50))
    address = Column(Text)
    region = Column(String(100))
    registration_date = Column(Date)
    status = Column(String(20))
    rating = Column(Float)
    created_at = Column(DateTime, default=datetime.utcnow)

    # Relationships
    transactions = relationship("Transaction", back_populates="vendor")
    complaints = relationship("Complaint", back_populates="vendor")
    metrics = relationship("VendorMetric", back_populates="vendor")

    def __repr__(self):
        return f"<Vendor(vendor_code='{self.vendor_code}', name='{self.vendor_name}')>"


class Transaction(Base):
    """Transactions Table"""
    __tablename__ = 'transactions'

    transaction_id = Column(Integer, primary_key=True, autoincrement=True)
    vendor_id = Column(Integer, ForeignKey('vendors.vendor_id'), nullable=False)
    transaction_date = Column(Date)
    transaction_amount = Column(Float)
    status = Column(String(50))
    response_time_hours = Column(Integer)
    success = Column(Boolean)

    # Relationship
    vendor = relationship("Vendor", back_populates="transactions")

    def __repr__(self):
        return f"<Transaction(id={self.transaction_id}, vendor_id={self.vendor_id})>"


class Complaint(Base):
    """Complaints Table"""
    __tablename__ = 'complaints'

    complaint_id = Column(Integer, primary_key=True, autoincrement=True)
    vendor_id = Column(Integer, ForeignKey('vendors.vendor_id'), nullable=False)
    complaint_date = Column(Date)
    complaint_type = Column(String(100))
    severity = Column(String(20))
    resolved = Column(Boolean, default=False)
    resolution_time_days = Column(Integer)

    # Relationship
    vendor = relationship("Vendor", back_populates="complaints")

    def __repr__(self):
        return f"<Complaint(id={self.complaint_id}, vendor_id={self.vendor_id})>"


class VendorMetric(Base):
    """Vendor Performance Metrics Table"""
    __tablename__ = 'vendor_metrics'

    metric_id = Column(Integer, primary_key=True, autoincrement=True)
    vendor_id = Column(Integer, ForeignKey('vendors.vendor_id'), nullable=False)
    metric_date = Column(Date)
    total_transactions = Column(Integer)
    successful_transactions = Column(Integer)
    total_complaints = Column(Integer)
    avg_response_time = Column(Float)
    avg_delivery_time = Column(Float)
    complaints_per_1000_transactions = Column(Float)

    # Relationship
    vendor = relationship("Vendor", back_populates="metrics")

    def __repr__(self):
        return f"<VendorMetric(vendor_id={self.vendor_id}, date={self.metric_date})>"
```

```python
%%writefile /content/drive/MyDrive/vendor-analysis-project/scraper/utils.py
"""
Scraper Utility Functions
Helper functions for web scraping
"""
import time
import logging
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import sys
sys.path.append('/content/drive/MyDrive/vendor-analysis-project')
from config.settings import SCRAPING_CONFIG

# Setup logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

def wait_for_element(driver, by, selector, timeout=10):
    """
    Wait for an element to be present

    Args:
        driver: Selenium WebDriver instance
        by: Selenium By locator (e.g., By.CSS_SELECTOR)
        selector: Element selector
        timeout: Maximum wait time in seconds

    Returns:
        WebElement if found, None otherwise
    """
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((by, selector))
        )
        return element
    except TimeoutException:
        logger.error(f"Timeout waiting for element: {selector}")
        return None

def wait_for_clickable(driver, by, selector, timeout=10):
    """
    Wait for an element to be clickable

    Args:
        driver: Selenium WebDriver instance
        by: Selenium By locator
        selector: Element selector
        timeout: Maximum wait time in seconds

    Returns:
        WebElement if found, None otherwise
    """
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.element_to_be_clickable((by, selector))
        )
        return element
    except TimeoutException:
        logger.error(f"Timeout waiting for clickable element: {selector}")
        return None

def safe_find_element(driver, by, selector):
    """
    Safely find an element without throwing exception

    Args:
        driver: Selenium WebDriver instance
        by: Selenium By locator
        selector: Element selector

    Returns:
        WebElement if found, None otherwise
    """
    try:
        return driver.find_element(by, selector)
    except NoSuchElementException:
        logger.warning(f"Element not found: {selector}")
        return None

def safe_click(driver, element, max_retries=3):
    """
    Safely click an element with retry logic

    Args:
        driver: Selenium WebDriver instance
        element: WebElement to click
        max_retries: Maximum number of retry attempts

    Returns:
        True if successful, False otherwise
    """
    for attempt in range(max_retries):
        try:
            element.click()
            return True
        except Exception as e:
            logger.warning(f"Click attempt {attempt + 1} failed: {str(e)}")
            time.sleep(1)
            if attempt == max_retries - 1:
                logger.error("All click attempts failed")
                return False
    return False

def extract_text(element, default=""):
    """
    Safely extract text from an element

    Args:
        element: WebElement
        default: Default value if extraction fails

    Returns:
        Extracted text or default value
    """
    try:
        return element.text.strip() if element else default
    except Exception as e:
        logger.warning(f"Text extraction failed: {str(e)}")
        return default

def retry_on_failure(func, max_retries=None, delay=None):
    """
    Decorator to retry function on failure

    Args:
        func: Function to execute
        max_retries: Maximum retry attempts
        delay: Delay between retries in seconds

    Returns:
        Function result or None if all retries fail
    """
    if max_retries is None:
        max_retries = SCRAPING_CONFIG['max_retries']
    if delay is None:
        delay = SCRAPING_CONFIG['retry_delay']

    def wrapper(*args, **kwargs):
        for attempt in range(max_retries):
            try:
                return func(*args, **kwargs)
            except Exception as e:
                logger.warning(f"Attempt {attempt + 1} failed: {str(e)}")
                if attempt < max_retries - 1:
                    time.sleep(delay)
                else:
                    logger.error(f"All {max_retries} attempts failed")
                    return None
        return None

    return wrapper

def scroll_to_element(driver, element):
    """
    Scroll to make element visible

    Args:
        driver: Selenium WebDriver instance
        element: WebElement to scroll to
    """
    try:
        driver.execute_script("arguments[0].scrollIntoView(true);", element)
        time.sleep(0.5)  # Brief pause after scrolling
    except Exception as e:
        logger.warning(f"Scroll failed: {str(e)}")
```

```python
%%writefile /content/drive/MyDrive/vendor-analysis-project/scraper/login_handler.py
"""
Login Handler
Handle authentication and session management
"""
import logging
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import sys
sys.path.append('/content/drive/MyDrive/vendor-analysis-project')
from config.settings import SELENIUM_CONFIG, WEBSITE_CONFIG, SELECTORS
from scraper.utils import wait_for_element, wait_for_clickable, safe_find_element

# Setup logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class LoginHandler:
    """Handle website authentication"""

    def __init__(self):
        self.driver = None
        self.is_logged_in = False

    def initialize_driver(self):
        """Initialize Selenium WebDriver with Chrome"""
        try:
            chrome_options = Options()

            # Colab-specific settings
            chrome_options.add_argument('--headless')
            chrome_options.add_argument('--no-sandbox')
            chrome_options.add_argument('--disable-dev-shm-usage')
            chrome_options.add_argument('--disable-gpu')
            chrome_options.add_argument('--window-size=1920,1080')
            chrome_options.add_argument('--disable-blink-features=AutomationControlled')
            chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
            chrome_options.add_experimental_option('useAutomationExtension', False)

            # Add user agent
            chrome_options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36')

            self.driver = webdriver.Chrome(options=chrome_options)
            self.driver.implicitly_wait(SELENIUM_CONFIG['implicit_wait'])
            self.driver.set_page_load_timeout(SELENIUM_CONFIG['page_load_timeout'])

            logger.info("✅ WebDriver initialized successfully")
            return self.driver

        except Exception as e:
            logger.error(f"❌ Failed to initialize WebDriver: {str(e)}")
            raise

    def login(self, email, password):
        """
        Login to the website

        Args:
            email: User email
            password: User password

        Returns:
            bool: True if login successful, False otherwise
        """
        try:
            if not self.driver:
                self.initialize_driver()

            logger.info(f"🔐 Attempting login to: {WEBSITE_CONFIG['login_url']}")

            # Navigate to login page
            self.driver.get(WEBSITE_CONFIG['login_url'])
            time.sleep(2)  # Wait for page load

            # Wait for and fill email field
            email_field = wait_for_element(
                self.driver,
                By.CSS_SELECTOR,
                SELECTORS['login']['email_field']
            )

            if not email_field:
                logger.error("❌ Email field not found")
                return False

            email_field.clear()
            email_field.send_keys(email)
            logger.info("✓ Email entered")

            # Wait for and fill password field
            password_field = wait_for_element(
                self.driver,
                By.CSS_SELECTOR,
                SELECTORS['login']['password_field']
            )

            if not password_field:
                logger.error("❌ Password field not found")
                return False

            password_field.clear()
            password_field.send_keys(password)
            logger.info("✓ Password entered")

            # Click login button
            login_button = wait_for_clickable(
                self.driver,
                By.CSS_SELECTOR,
                SELECTORS['login']['login_button']
            )

            if not login_button:
                logger.error("❌ Login button not found")
                return False

            login_button.click()
            logger.info("✓ Login button clicked")

            # Wait for successful login (check for dashboard or success indicator)
            time.sleep(3)

            # Verify login success
            success_indicator = wait_for_element(
                self.driver,
                By.CSS_SELECTOR,
                SELECTORS['login']['success_indicator'],
                timeout=10
            )

            if success_indicator:
                self.is_logged_in = True
                logger.info("✅ Login successful!")
                return True
            else:
                logger.error("❌ Login failed - success indicator not found")
                return False

        except Exception as e:
            logger.error(f"❌ Login error: {str(e)}")
            return False

    def check_login_status(self):
        """Check if currently logged in"""
        return self.is_logged_in

    def navigate_to_vendors_page(self):
        """Navigate to vendors master page"""
        try:
            if not self.is_logged_in:
                logger.error("❌ Cannot navigate - not logged in")
                return False

            logger.info(f"📄 Navigating to vendors page: {WEBSITE_CONFIG['vendors_url']}")
            self.driver.get(WEBSITE_CONFIG['vendors_url'])
            time.sleep(2)

            # Verify vendors page loaded
            vendors_table = wait_for_element(
                self.driver,
                By.CSS_SELECTOR,
                SELECTORS['vendors']['vendor_table'],
                timeout=10
            )

            if vendors_table:
                logger.info("✅ Vendors page loaded successfully")
                return True
            else:
                logger.error("❌ Vendors table not found")
                return False

        except Exception as e:
            logger.error(f"❌ Navigation error: {str(e)}")
            return False

    def close(self):
        """Close the browser and cleanup"""
        if self.driver:
            self.driver.quit()
            logger.info("🔒 Browser closed")
            self.is_logged_in = False

    def __enter__(self):
        """Context manager entry"""
        self.initialize_driver()
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        """Context manager exit"""
        self.close()
```

```python
%%writefile /content/drive/MyDrive/vendor-analysis-project/scraper/vendor_scraper.py
"""
Vendor Scraper
Main scraping logic for extracting vendor data
"""
import logging
import time
import pandas as pd
from datetime import datetime
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import sys
sys.path.append('/content/drive/MyDrive/vendor-analysis-project')
from config.settings import SCRAPING_CONFIG, PATHS, SELECTORS
from scraper.login_handler import LoginHandler
from scraper.utils import wait_for_element, safe_click, extract_text, scroll_to_element

# Setup logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class VendorScraper:
    """Scrape vendor data from the website"""

    def __init__(self, email, password):
        self.email = email
        self.password = password
        self.login_handler = LoginHandler()
        self.vendors_data = []
        self.transactions_data = []
        self.complaints_data = []
        self.scraped_count = 0

    def initialize(self):
        """Initialize scraper and login"""
        try:
            # Initialize driver
            self.login_handler.initialize_driver()

            # Login
            if self.login_handler.login(self.email, self.password):
                logger.info("✅ Scraper initialized and logged in")
                return True
            else:
                logger.error("❌ Failed to login")
                return False

        except Exception as e:
            logger.error(f"❌ Initialization error: {str(e)}")
            return False

    def get_vendor_rows(self):
        """Get all vendor rows from the table"""
        try:
            # Navigate to vendors page
            if not self.login_handler.navigate_to_vendors_page():
                return []

            # Find all vendor rows
            vendor_rows = self.login_handler.driver.find_elements(
                By.CSS_SELECTOR,
                SELECTORS['vendors']['vendor_rows']
            )

            logger.info(f"📊 Found {len(vendor_rows)} vendor rows")
            return vendor_rows

        except Exception as e:
            logger.error(f"❌ Error getting vendor rows: {str(e)}")
            return []

    def extract_vendor_details(self, vendor_row, index):
        """
        Extract vendor details from a row

        Args:
            vendor_row: WebElement representing vendor row
            index: Row index

        Returns:
            dict: Vendor data
        """
        try:
            vendor_data = {
                'vendor_code': '',
                'vendor_name': '',
                'email': '',
                'phone': '',
                'address': '',
                'region': '',
                'registration_date': None,
                'status': '',
                'rating': None,
                'total_transactions': 0,
                'successful_transactions': 0,
                'total_complaints': 0,
                'avg_response_time': None,
                'avg_delivery_time': None,
                'scraped_at': datetime.now()
            }

            # Scroll to row
            scroll_to_element(self.login_handler.driver, vendor_row)
            time.sleep(0.5)

            # Extract data from table cells
            cells = vendor_row.find_elements(By.TAG_NAME, 'td')

            if len(cells) >= 3:
                vendor_data['vendor_code'] = extract_text(cells[0])
                vendor_data['vendor_name'] = extract_text(cells[1])
                vendor_data['region'] = extract_text(cells[2]) if len(cells) > 2 else ''
                vendor_data['status'] = extract_text(cells[3]) if len(cells) > 3 else 'Active'

            # Try to click "View" button to get more details
            view_button = vendor_row.find_element(By.CSS_SELECTOR, SELECTORS['vendors']['view_button'])

            if view_button and safe_click(self.login_handler.driver, view_button):
                time.sleep(2)  # Wait for details page to load

                # Extract detailed information
                vendor_data = self.extract_detailed_info(vendor_data)

                # Go back to list
                self.login_handler.driver.back()
                time.sleep(2)

            logger.info(f"✓ Extracted vendor {index + 1}: {vendor_data['vendor_name']}")
            return vendor_data

        except StaleElementReferenceException:
            logger.warning(f"⚠ Stale element at index {index}, retrying...")
            return None
        except Exception as e:
            logger.error(f"❌ Error extracting vendor {index}: {str(e)}")
            return None

    def extract_detailed_info(self, vendor_data):
        """
        Extract detailed information from vendor detail page

        Args:
            vendor_data: Existing vendor data dict

        Returns:
            dict: Updated vendor data
        """
        try:
            driver = self.login_handler.driver

            # Example selectors - UPDATE THESE BASED ON YOUR ACTUAL WEBSITE
            selectors_map = {
                'email': '#vendor-email',
                'phone': '#vendor-phone',
                'address': '#vendor-address',
                'registration_date': '#reg-date',
                'rating': '#vendor-rating',
                'total_transactions': '#total-transactions',
                'successful_transactions': '#successful-transactions',
                'total_complaints': '#total-complaints',
                'avg_response_time': '#avg-response-time',
                'avg_delivery_time': '#avg-delivery-time',
            }

            for field, selector in selectors_map.items():
                element = safe_find_element(driver, By.CSS_SELECTOR, selector)
                if element:
                    text = extract_text(element)

                    # Parse based on field type
                    if field in ['total_transactions', 'successful_transactions', 'total_complaints']:
                        try:
                            vendor_data[field] = int(text.replace(',', ''))
                        except:
                            vendor_data[field] = 0
                    elif field in ['rating', 'avg_response_time', 'avg_delivery_time']:
                        try:
                            vendor_data[field] = float(text)
                        except:
                            vendor_data[field] = None
                    elif field == 'registration_date':
                        vendor_data[field] = text  # Will parse later
                    else:
                        vendor_data[field] = text

            return vendor_data

        except Exception as e:
            logger.warning(f"⚠ Error extracting detailed info: {str(e)}")
            return vendor_data

    def scrape_all_vendors(self, max_vendors=None):
        """
        Scrape all vendors from the website

        Args:
            max_vendors: Maximum number of vendors to scrape (None for all)

        Returns:
            list: List of vendor data dictionaries
        """
        try:
            if not self.initialize():
                return []

            logger.info("🚀 Starting vendor scraping...")

            vendor_rows = self.get_vendor_rows()

            if not vendor_rows:
                logger.warning("⚠ No vendor rows found")
                return []

            total_to_scrape = min(len(vendor_rows), max_vendors) if max_vendors else len(vendor_rows)
            logger.info(f"📊 Will scrape {total_to_scrape} vendors")

            for index in range(total_to_scrape):
                try:
                    # Re-fetch rows to avoid stale element issues
                    vendor_rows = self.get_vendor_rows()

                    if index >= len(vendor_rows):
                        logger.warning(f"⚠ Index {index} out of range, stopping")
                        break

                    vendor_data = self.extract_vendor_details(vendor_rows[index], index)

                    if vendor_data:
                        self.vendors_data.append(vendor_data)
                        self.scraped_count += 1

                        # Save periodically
                        if self.scraped_count % SCRAPING_CONFIG['save_interval'] == 0:
                            self.save_data_incremental()
                            logger.info(f"💾 Progress saved: {self.scraped_count} vendors")

                    # Rate limiting
                    time.sleep(1)

                except Exception as e:
                    logger.error(f"❌ Error scraping vendor {index}: {str(e)}")
                    continue

            # Final save
            self.save_data_incremental()

            logger.info(f"✅ Scraping complete! Total vendors scraped: {self.scraped_count}")
            return self.vendors_data

        except Exception as e:
            logger.error(f"❌ Scraping error: {str(e)}")
            return self.vendors_data
        finally:
            self.cleanup()

    def save_data_incremental(self):
        """Save scraped data incrementally"""
        try:
            if not self.vendors_data:
                return

            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

            # Save vendors data
            vendors_df = pd.DataFrame(self.vendors_data)
            vendors_file = f"{PATHS['raw_data']}/vendors_raw_{timestamp}.csv"
            vendors_df.to_csv(vendors_file, index=False)

            logger.info(f"💾 Saved {len(self.vendors_data)} vendors to {vendors_file}")

        except Exception as e:
            logger.error(f"❌ Error saving data: {str(e)}")

    def cleanup(self):
        """Cleanup resources"""
        if self.login_handler:
            self.login_handler.close()
        logger.info("🧹 Cleanup complete")

def safe_find_element(driver, by, selector):
    """Helper function to safely find element"""
    try:
        return driver.find_element(by, selector)
    except NoSuchElementException:
        return None
```

```python
%%writefile /content/drive/MyDrive/vendor-analysis-project/database/queries.py
"""
Database Queries
SQL queries for data insertion and retrieval
"""
import logging
from sqlalchemy import func
from datetime import datetime
import sys
sys.path.append('/content/drive/MyDrive/vendor-analysis-project')
from database.models import Vendor, Transaction, Complaint, VendorMetric
from database.db_connection import db_manager

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class VendorQueries:
    """Handle vendor-related database operations"""

    @staticmethod
    def insert_vendor(vendor_data):
        """
        Insert a new vendor

        Args:
            vendor_data: Dictionary with vendor information

        Returns:
            Vendor object or None
        """
        session = db_manager.get_session()
        try:
            # Check if vendor already exists
            existing = session.query(Vendor).filter_by(
                vendor_code=vendor_data.get('vendor_code')
            ).first()

            if existing:
                logger.warning(f"Vendor {vendor_data.get('vendor_code')} already exists")
                return existing

            # Create new vendor
            vendor = Vendor(
                vendor_code=vendor_data.get('vendor_code'),
                vendor_name=vendor_data.get('vendor_name'),
                email=vendor_data.get('email'),
                phone=vendor_data.get('phone'),
                address=vendor_data.get('address'),
                region=vendor_data.get('region'),
                registration_date=vendor_data.get('registration_date'),
                status=vendor_data.get('status', 'Active'),
                rating=vendor_data.get('rating')
            )

            session.add(vendor)
            session.commit()
            logger.info(f"✓ Inserted vendor: {vendor.vendor_name}")
            return vendor

        except Exception as e:
            session.rollback()
            logger.error(f"❌ Error inserting vendor: {str(e)}")
            return None
        finally:
            db_manager.close_session(session)

    @staticmethod
    def bulk_insert_vendors(vendors_list):
        """
        Bulk insert vendors

        Args:
            vendors_list: List of vendor data dictionaries

        Returns:
            int: Number of vendors inserted
        """
        session = db_manager.get_session()
        inserted_count = 0

        try:
            for vendor_data in vendors_list:
                # Check if exists
                existing = session.query(Vendor).filter_by(
                    vendor_code=vendor_data.get('vendor_code')
                ).first()

                if not existing:
                    vendor = Vendor(**{
                        k: v for k, v in vendor_data.items()
                        if k in ['vendor_code', 'vendor_name', 'email', 'phone',
                                'address', 'region', 'registration_date', 'status', 'rating']
                    })
                    session.add(vendor)
                    inserted_count += 1

            session.commit()
            logger.info(f"✅ Bulk inserted {inserted_count} vendors")
            return inserted_count

        except Exception as e:
            session.rollback()
            logger.error(f"❌ Bulk insert error: {str(e)}")
            return inserted_count
        finally:
            db_manager.close_session(session)

    @staticmethod
    def get_all_vendors():
        """Get all vendors"""
        session = db_manager.get_session()
        try:
            vendors = session.query(Vendor).all()
            return vendors
        finally:
            db_manager.close_session(session)

    @staticmethod
    def get_vendor_by_code(vendor_code):
        """Get vendor by code"""
        session = db_manager.get_session()
        try:
            vendor = session.query(Vendor).filter_by(vendor_code=vendor_code).first()
            return vendor
        finally:
            db_manager.close_session(session)

    @staticmethod
    def update_vendor_metrics(vendor_id, metrics_data):
        """
        Update vendor metrics

        Args:
            vendor_id: Vendor ID
            metrics_data: Dictionary with metrics
        """
        session = db_manager.get_session()
        try:
            metric = VendorMetric(
                vendor_id=vendor_id,
                metric_date=datetime.now().date(),
                **metrics_data
            )
            session.add(metric)
            session.commit()
            logger.info(f"✓ Updated metrics for vendor {vendor_id}")
        except Exception as e:
            session.rollback()
            logger.error(f"❌ Error updating metrics: {str(e)}")
        finally:
            db_manager.close_session(session)

    @staticmethod
    def get_vendor_statistics():
        """Get overall vendor statistics"""
        session = db_manager.get_session()
        try:
            stats = {
                'total_vendors': session.query(func.count(Vendor.vendor_id)).scalar(),
                'active_vendors': session.query(func.count(Vendor.vendor_id)).filter(
                    Vendor.status == 'Active'
                ).scalar(),
                'avg_rating': session.query(func.avg(Vendor.rating)).scalar(),
                'total_transactions': session.query(func.count(Transaction.transaction_id)).scalar(),
                'total_complaints': session.query(func.count(Complaint.complaint_id)).scalar(),
            }
            return stats
        finally:
            db_manager.close_session(session)
```

```python
%%writefile /content/drive/MyDrive/vendor-analysis-project/scraper/selector_finder.py
"""
Selector Finder Tool
Help identify CSS selectors from the target website
"""
import logging
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class SelectorFinder:
    """Tool to help find CSS selectors on the website"""

    def __init__(self, url):
        self.url = url
        self.driver = None

    def initialize_driver(self):
        """Initialize Chrome driver"""
        chrome_options = Options()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        chrome_options.add_argument('--window-size=1920,1080')

        self.driver = webdriver.Chrome(options=chrome_options)
        logger.info("✅ Driver initialized")

    def load_page(self):
        """Load the target page"""
        self.driver.get(self.url)
        time.sleep(3)
        logger.info(f"✅ Loaded page: {self.url}")

    def find_login_elements(self):
        """Find login page elements"""
        print("\n🔍 Searching for LOGIN elements...")
        print("=" * 60)

        # Common email field patterns
        email_patterns = [
            ('ID: email', By.ID, 'email'),
            ('ID: username', By.ID, 'username'),
            ('ID: user_email', By.ID, 'user_email'),
            ('Name: email', By.NAME, 'email'),
            ('Name: username', By.NAME, 'username'),
            ('Type: email', By.CSS_SELECTOR, 'input[type="email"]'),
            ('Type: text (email)', By.CSS_SELECTOR, 'input[type="text"]'),
        ]

        # Common password field patterns
        password_patterns = [
            ('ID: password', By.ID, 'password'),
            ('ID: user_password', By.ID, 'user_password'),
            ('Name: password', By.NAME, 'password'),
            ('Type: password', By.CSS_SELECTOR, 'input[type="password"]'),
        ]

        # Common button patterns
        button_patterns = [
            ('ID: login-btn', By.ID, 'login-btn'),
            ('ID: submit', By.ID, 'submit'),
            ('Type: submit', By.CSS_SELECTOR, 'button[type="submit"]'),
            ('Class: btn-login', By.CLASS_NAME, 'btn-login'),
            ('Text: Login', By.XPATH, "//button[contains(text(), 'Login')]"),
            ('Text: Sign In', By.XPATH, "//button[contains(text(), 'Sign In')]"),
        ]

        found = {}

        # Find email field
        print("\n📧 EMAIL FIELD:")
        for name, by, selector in email_patterns:
            try:
                element = self.driver.find_element(by, selector)
                print(f"  ✅ Found: {name} → Selector: '{selector}'")
                if 'email' not in found:
                    found['email'] = selector
                break
            except:
                pass

        # Find password field
        print("\n🔒 PASSWORD FIELD:")
        for name, by, selector in password_patterns:
            try:
                element = self.driver.find_element(by, selector)
                print(f"  ✅ Found: {name} → Selector: '{selector}'")
                if 'password' not in found:
                    found['password'] = selector
                break
            except:
                pass

        # Find login button
        print("\n🔘 LOGIN BUTTON:")
        for name, by, selector in button_patterns:
            try:
                element = self.driver.find_element(by, selector)
                print(f"  ✅ Found: {name} → Selector: '{selector}'")
                if 'button' not in found:
                    found['button'] = selector
                break
            except:
                pass

        return found

    def get_page_source_preview(self):
        """Get a preview of the page source"""
        source = self.driver.page_source
        print("\n📄 PAGE SOURCE PREVIEW (first 1000 chars):")
        print("=" * 60)
        print(source[:1000])
        return source

    def find_all_inputs(self):
        """Find all input fields on the page"""
        print("\n🔍 ALL INPUT FIELDS:")
        print("=" * 60)

        inputs = self.driver.find_elements(By.TAG_NAME, 'input')
        for i, inp in enumerate(inputs, 1):
            input_type = inp.get_attribute('type')
            input_id = inp.get_attribute('id')
            input_name = inp.get_attribute('name')
            input_class = inp.get_attribute('class')
            placeholder = inp.get_attribute('placeholder')

            print(f"\nInput #{i}:")
            print(f"  Type: {input_type}")
            print(f"  ID: {input_id}")
            print(f"  Name: {input_name}")
            print(f"  Class: {input_class}")
            print(f"  Placeholder: {placeholder}")

    def find_all_buttons(self):
        """Find all buttons on the page"""
        print("\n🔍 ALL BUTTONS:")
        print("=" * 60)

        buttons = self.driver.find_elements(By.TAG_NAME, 'button')
        for i, btn in enumerate(buttons, 1):
            btn_type = btn.get_attribute('type')
            btn_id = btn.get_attribute('id')
            btn_class = btn.get_attribute('class')
            btn_text = btn.text

            print(f"\nButton #{i}:")
            print(f"  Type: {btn_type}")
            print(f"  ID: {btn_id}")
            print(f"  Class: {btn_class}")
            print(f"  Text: {btn_text}")

    def close(self):
        """Close the driver"""
        if self.driver:
            self.driver.quit()

# Main function to run selector finder
def find_selectors(url):
    """
    Find selectors on a webpage

    Args:
        url: URL to inspect
    """
    finder = SelectorFinder(url)

    try:
        finder.initialize_driver()
        finder.load_page()

        # Find login elements
        found_selectors = finder.find_login_elements()

        # Show all inputs and buttons
        print("\n" + "=" * 60)
        finder.find_all_inputs()
        finder.find_all_buttons()

        # Show page source preview
        # finder.get_page_source_preview()

        return found_selectors

    except Exception as e:
        print(f"\n❌ Error: {str(e)}")
        return {}
    finally:
        finder.close()
```

```python
# ============================================================================
# COMPLETE BEDROCK VENDOR SCRAPER - ALL PAGES WITH FULL DETAILS
# Scrapes all 17,372 vendors from all 174 pages + clicks View for details
# ============================================================================

import sys
sys.path.append('/content/drive/MyDrive/vendor-analysis-project')

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from datetime import datetime
import os
import json

print("\n" + "=" * 100)
print("COMPLETE BEDROCK VENDOR SCRAPER - ALL PAGES + FULL DETAILS")
print("=" * 100 + "\n")

# ============================================================================
# CONFIGURATION
# ============================================================================
EMAIL = 'Ramaalfaran82@gmail.com'
PASSWORD = 'Ramaalfaran82@gmail.com'
LOGIN_URL = 'https://qa.mybedrock.com/admin'
VENDOR_MASTER_URL = 'https://qa.mybedrock.com/admin/clients/vendor_master'

# Save settings
SAVE_EVERY_N_VENDORS = 50  # Save progress after every 50 vendors
DATA_DIR = '/content/drive/MyDrive/vendor-analysis-project/data/raw'
os.makedirs(DATA_DIR, exist_ok=True)

print("📋 CONFIGURATION:")
print(f"   Target: All 17,372 vendors from ~174 pages")
print(f"   Mode: Full details (Table + View page)")
print(f"   Save every: {SAVE_EVERY_N_VENDORS} vendors")
print(f"   Data directory: {DATA_DIR}\n")

# ============================================================================
# INITIALIZE WEBDRIVER
# ============================================================================
print("🔧 Initializing WebDriver...")
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--window-size=1920,1080')
chrome_options.add_argument('--disable-blink-features=AutomationControlled')
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)
chrome_options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36')

driver = webdriver.Chrome(options=chrome_options)
driver.implicitly_wait(10)
driver.set_page_load_timeout(30)
wait = WebDriverWait(driver, 15)
print("   ✓ WebDriver ready\n")

# ============================================================================
# LOGIN
# ============================================================================
print("🔐 Logging in...")
try:
    driver.get(LOGIN_URL)
    time.sleep(3)
    driver.find_element(By.ID, 'email').send_keys(EMAIL)
    driver.find_element(By.ID, 'password').send_keys(PASSWORD)
    driver.find_element(By.CSS_SELECTOR, 'button.btn.btn-info.btn-block').click()
    time.sleep(5)

    if 'authentication' not in driver.current_url:
        print("   ✓ Login successful\n")
    else:
        print("   ✗ Login failed")
        driver.quit()
        sys.exit(1)
except Exception as e:
    print(f"   ✗ Login error: {str(e)}")
    driver.quit()
    sys.exit(1)

# ============================================================================
# SETUP VENDOR PAGE
# ============================================================================
print("⚙️ Setting up vendor master page...")
driver.get(VENDOR_MASTER_URL)
time.sleep(3)

# Set page size to 100
try:
    page_size_dropdown = driver.find_element(By.CSS_SELECTOR, 'select.form-control')
    driver.execute_script("arguments[0].value = '100';", page_size_dropdown)
    driver.execute_script("arguments[0].dispatchEvent(new Event('change'));", page_size_dropdown)
    time.sleep(3)
    print("   ✓ Page size set to 100\n")
except Exception as e:
    print(f"   ⚠ Could not set page size: {str(e)}\n")

# ============================================================================
# EXTRACTION FUNCTIONS
# ============================================================================

def extract_table_data(row):
    """Extract data from table row"""
    try:
        cells = row.find_elements(By.TAG_NAME, 'td')
        texts = [c.text.strip() for c in cells]

        # Get vendor name (keep original with View/Details/Log)
        vendor_name_raw = texts[3] if len(texts) > 3 else ''

        data = {
            # Table data (22 columns)
            'bedrock_id': texts[1] if len(texts) > 1 else '',
            'vendor_number': texts[2] if len(texts) > 2 else '',
            'vendor_name_raw': vendor_name_raw,  # Keep original
            'dba_name': texts[4] if len(texts) > 4 else '',
            'onboarding_status': texts[5] if len(texts) > 5 else '',
            'audit_status': texts[6] if len(texts) > 6 else '',
            'primary_contact': texts[7] if len(texts) > 7 else '',
            'primary_email': texts[8] if len(texts) > 8 else '',
            'tax_id': texts[9] if len(texts) > 9 else '',
            'audit_period': texts[10] if len(texts) > 10 else '',
            'phone': texts[11] if len(texts) > 11 else '',
            'total_spend': texts[12] if len(texts) > 12 else '',
            'spend_2024': texts[13] if len(texts) > 13 else '',
            'spend_2023': texts[14] if len(texts) > 14 else '',
            'spend_2022': texts[15] if len(texts) > 15 else '',
            'legacy_vendor_id': texts[16] if len(texts) > 16 else '',
            'login_status': texts[17] if len(texts) > 17 else '',
            'modified_by': texts[18] if len(texts) > 18 else '',
            'modified_date': texts[19] if len(texts) > 19 else '',
            'modified_type': texts[20] if len(texts) > 20 else '',
            'created_by': texts[21] if len(texts) > 21 else '',
        }

        return data
    except Exception as e:
        print(f"      ✗ Error extracting table data: {e}")
        return None

def extract_detail_page_data():
    """Extract additional data from detail page after clicking View"""
    try:
        time.sleep(2)  # Wait for page to load

        detail_data = {
            'vendor_type': '',
            'payment_terms': '',
            'estimated_annual_spend': '',
            'one_time_vendor': '',
            'registered_company_name': '',
            'company_dba_name': '',
            'registered_address_1': '',
            'registered_address_2': '',
            'city': '',
            'county': '',
            'state': '',
            'postal_code': '',
            'country': '',
            'business_entity_type': '',
            'year_founded': '',
            'duns_number': '',
            'primary_contact_first_name': '',
            'primary_contact_last_name': '',
            'primary_contact_email': '',
            'primary_contact_phone': '',
            'website_url': '',
            'compliance_verification': '',
            'banking_verification': '',
        }

        # Get page text for text-based extraction
        page_text = driver.find_element(By.TAG_NAME, 'body').text
        lines = page_text.split('\n')

        # Text-based field extraction
        text_fields = {
            'Payment Terms :': 'payment_terms',
            'Vendor Type :': 'vendor_type',
            'Estimated Annual Spend :': 'estimated_annual_spend',
            'Is this a one-time vendor? :': 'one_time_vendor',
            'Global Compliance Verification :': 'compliance_verification',
            'Banking Information Verification :': 'banking_verification',
            'Business Entity Type :': 'business_entity_type',
            'Year Founded :': 'year_founded',
            'County :': 'county',
        }

        for label, field_name in text_fields.items():
            for line in lines:
                if label in line:
                    try:
                        value = line.split(label, 1)[1].strip()
                        detail_data[field_name] = value
                        break
                    except:
                        pass

        # Extract from input fields
        input_mappings = [
            ('input[name*="registered_company_name"]', 'registered_company_name'),
            ('input[name*="company_dba"]', 'company_dba_name'),
            ('input[name*="address"]', 'registered_address_1'),
            ('input[name*="city"]', 'city'),
            ('input[name*="postal"]', 'postal_code'),
            ('input[name*="duns"]', 'duns_number'),
            ('input[name*="website"]', 'website_url'),
            ('input[name*="phone"]', 'primary_contact_phone'),
        ]

        for selector, field_name in input_mappings:
            try:
                elements = driver.find_elements(By.CSS_SELECTOR, selector)
                if elements:
                    value = elements[0].get_attribute('value')
                    if value:
                        detail_data[field_name] = value
            except:
                pass

        # State (can be select or input)
        try:
            state_elem = driver.find_element(By.CSS_SELECTOR, 'select[name*="state"], input[name*="state"]')
            detail_data['state'] = state_elem.get_attribute('value')
        except:
            pass

        # Country (can be select or input)
        try:
            country_elem = driver.find_element(By.CSS_SELECTOR, 'select[name*="country"], input[name*="country"]')
            detail_data['country'] = country_elem.get_attribute('value')
        except:
            pass

        return detail_data

    except Exception as e:
        print(f"      ✗ Error extracting detail page: {e}")
        return None

def go_to_next_page():
    """Navigate to next page"""
    try:
        time.sleep(1)

        # Find next button (not disabled)
        next_button = driver.find_element(By.CSS_SELECTOR, 'li.paginate_button.next:not(.disabled) a')

        # Check if parent is disabled
        parent_li = next_button.find_element(By.XPATH, '..')
        if 'disabled' in parent_li.get_attribute('class'):
            return False

        # Click using JavaScript
        driver.execute_script("arguments[0].click();", next_button)
        time.sleep(3)

        # Wait for table to reload
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'tbody tr')))
        return True

    except Exception as e:
        return False

def save_progress(data, total_count):
    """Save progress to CSV"""
    if not data:
        return None

    df = pd.DataFrame(data)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filepath = f"{DATA_DIR}/vendors_progress_{total_count}vendors_{timestamp}.csv"
    df.to_csv(filepath, index=False)
    return filepath

# ============================================================================
# MAIN SCRAPING LOOP
# ============================================================================
print("=" * 100)
print("🚀 STARTING COMPLETE SCRAPING (ALL PAGES + DETAILS)")
print("=" * 100 + "\n")

all_vendors_data = []
page_count = 1
total_scraped = 0
total_with_details = 0
start_time = time.time()

try:
    while True:
        print(f"\n{'='*100}")
        print(f"PAGE {page_count}")
        print(f"{'='*100}")

        # Wait for table to load
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'tbody tr')))
        time.sleep(2)

        # Get all rows on current page
        rows = driver.find_elements(By.CSS_SELECTOR, 'tbody tr')

        # Check for "No data" message
        if len(rows) == 1 and ('No data available' in rows[0].text or 'No matching records' in rows[0].text):
            print(f"   ⚠ No data on this page - stopping")
            break

        print(f"   Found {len(rows)} vendors on this page\n")

        # Process each vendor on this page
        for idx in range(len(rows)):
            try:
                # Navigate back to vendor master (in case we clicked away)
                driver.get(VENDOR_MASTER_URL)
                time.sleep(2)

                # Re-fetch rows (important for stale element prevention)
                wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'tbody tr')))
                rows = driver.find_elements(By.CSS_SELECTOR, 'tbody tr')

                if idx >= len(rows):
                    print(f"   ⚠ Row {idx} out of range")
                    break

                row = rows[idx]
                vendor_num = total_scraped + 1

                print(f"   [{vendor_num}] Processing vendor {idx + 1}/{len(rows)} on page {page_count}...")

                # STEP 1: Extract table data
                vendor_data = extract_table_data(row)

                if not vendor_data:
                    print(f"      ✗ Failed to extract table data")
                    continue

                # Add metadata
                vendor_data['page_number'] = page_count
                vendor_data['row_index_on_page'] = idx + 1
                vendor_data['scraped_at'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

                # STEP 2: Click View and extract detail page data
                try:
                    cells = row.find_elements(By.TAG_NAME, 'td')
                    if len(cells) > 3:
                        vendor_name_cell = cells[3]
                        view_link = vendor_name_cell.find_element(By.TAG_NAME, 'a')

                        # Click View
                        driver.execute_script("arguments[0].click();", view_link)
                        time.sleep(2)

                        # Extract detail page data
                        detail_data = extract_detail_page_data()

                        if detail_data:
                            # Merge detail data into vendor data
                            vendor_data.update(detail_data)
                            total_with_details += 1

                            # Show some key extracted fields
                            extracted_fields = []
                            if detail_data.get('payment_terms'):
                                extracted_fields.append(f"Terms: {detail_data['payment_terms'][:20]}")
                            if detail_data.get('city'):
                                extracted_fields.append(f"City: {detail_data['city']}")
                            if detail_data.get('vendor_type'):
                                extracted_fields.append(f"Type: {detail_data['vendor_type'][:20]}")

                            if extracted_fields:
                                print(f"      ✓ Details: {' | '.join(extracted_fields)}")
                            else:
                                print(f"      ✓ Detail page scraped")
                        else:
                            print(f"      ⚠ Could not extract detail data")
                    else:
                        print(f"      ⚠ Could not find View link")

                except Exception as e:
                    print(f"      ⚠ Error getting details: {str(e)}")

                # Add to list
                all_vendors_data.append(vendor_data)
                total_scraped += 1

                # Save progress periodically
                if total_scraped % SAVE_EVERY_N_VENDORS == 0:
                    filepath = save_progress(all_vendors_data, total_scraped)

                    elapsed = time.time() - start_time
                    rate = total_scraped / elapsed if elapsed > 0 else 0
                    remaining = 17372 - total_scraped
                    eta_minutes = (remaining / rate / 60) if rate > 0 else 0

                    print(f"\n   💾 Progress saved: {filepath}")
                    print(f"   📊 Total scraped: {total_scraped} vendors")
                    print(f"   📋 With details: {total_with_details} ({total_with_details/total_scraped*100:.1f}%)")
                    print(f"   ⏱️  Speed: {rate:.2f} vendors/sec | ETA: {eta_minutes:.1f} min\n")

                # Small delay between vendors
                time.sleep(0.5)

            except KeyboardInterrupt:
                raise  # Re-raise to catch in outer try
            except Exception as e:
                print(f"   ✗ Error with vendor {idx + 1}: {str(e)}")
                continue

        # All vendors on this page processed
        print(f"\n   ✅ Page {page_count} complete - {len(rows)} vendors processed")
        print(f"   📊 Running total: {total_scraped} vendors\n")

        # Try to go to next page
        print(f"   → Navigating to page {page_count + 1}...")
        if not go_to_next_page():
            print(f"\n   ✅ Last page reached!")
            break

        page_count += 1
        time.sleep(1)

except KeyboardInterrupt:
    print("\n\n⚠️  SCRAPING INTERRUPTED BY USER")
except Exception as e:
    print(f"\n✗ Unexpected error: {str(e)}")
    import traceback
    traceback.print_exc()

# ============================================================================
# SAVE FINAL DATA
# ============================================================================
print("\n" + "=" * 100)
print("💾 SAVING FINAL DATA")
print("=" * 100 + "\n")

if all_vendors_data:
    df = pd.DataFrame(all_vendors_data)

    # Remove duplicates based on bedrock_id
    original_count = len(df)
    df = df.drop_duplicates(subset=['bedrock_id'], keep='first')
    duplicates = original_count - len(df)

    # Save final CSV
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    final_csv = f'{DATA_DIR}/vendors_complete_with_details_{timestamp}.csv'
    df.to_csv(final_csv, index=False)

    print(f"✅ FINAL DATA SAVED: {final_csv}\n")

    # Calculate statistics
    total_time = time.time() - start_time
    hours = int(total_time // 3600)
    minutes = int((total_time % 3600) // 60)

    print(f"📊 FINAL STATISTICS:")
    print(f"   {'='*70}")
    print(f"   Pages scraped: {page_count}")
    print(f"   Total records: {original_count}")
    print(f"   Duplicates removed: {duplicates}")
    print(f"   Unique vendors: {len(df)}")
    print(f"   With detail data: {total_with_details} ({total_with_details/len(df)*100:.1f}%)")
    print(f"   Total time: {hours}h {minutes}m")
    print(f"   Average speed: {total_scraped/total_time:.2f} vendors/second\n")

    # Data completeness
    print(f"   📋 DATA COMPLETENESS:")
    print(f"   Primary Email: {df['primary_email'].notna().sum()} ({df['primary_email'].notna().sum()/len(df)*100:.1f}%)")
    print(f"   Phone: {df['phone'].notna().sum()} ({df['phone'].notna().sum()/len(df)*100:.1f}%)")
    print(f"   Tax ID: {df['tax_id'].notna().sum()} ({df['tax_id'].notna().sum()/len(df)*100:.1f}%)")
    print(f"   City: {df['city'].notna().sum()} ({df['city'].notna().sum()/len(df)*100:.1f}%)")
    print(f"   Payment Terms: {df['payment_terms'].notna().sum()} ({df['payment_terms'].notna().sum()/len(df)*100:.1f}%)")
    print(f"   Vendor Type: {df['vendor_type'].notna().sum()} ({df['vendor_type'].notna().sum()/len(df)*100:.1f}%)\n")

    # Top vendors by spend
    print(f"   💰 TOP 10 VENDORS BY TOTAL SPEND:")
    df['total_spend_clean'] = df['total_spend'].replace(r'[\$,]', '', regex=True).replace('', '0')
    df['total_spend_clean'] = pd.to_numeric(df['total_spend_clean'], errors='coerce').fillna(0)

    top_vendors = df.nlargest(10, 'total_spend_clean')[['vendor_name_raw', 'total_spend', 'primary_email']]
    for idx, row in top_vendors.iterrows():
        vendor_name = row['vendor_name_raw'].split('\n')[0][:40]
        print(f"   {vendor_name:40s} ${row['total_spend']:>15s}")

    # Sample data
    print(f"\n   📋 SAMPLE DATA (First 5 vendors):")
    sample_cols = ['bedrock_id', 'vendor_name_raw', 'primary_email', 'city', 'payment_terms']
    available_cols = [col for col in sample_cols if col in df.columns]
    print(df[available_cols].head(5).to_string(index=False))

else:
    print("✗ No data to save")

# Cleanup
print("\n" + "=" * 100)
driver.quit()
print("✓ Browser closed")
print("=" * 100)
print(f"\n✅ SCRAPING COMPLETE")
print(f"   Total Vendors: {total_scraped}")
print(f"   Pages Processed: {page_count}")
print(f"   With Full Details: {total_with_details}")
print("=" * 100 + "\n")
```
